### introduction to matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
x
```

```python
y
```

```python
# functional

plt.plot(x,y, 'r-')
# plt.show() # remove a linha de texto acima do gráfico
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# adicionar subplot

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # distance from right, distance from bottom, width, heigth

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)


axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

### figure size and DPI

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
# two curves in the same figure

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'x squared')
ax.plot(x,x**3, label = 'x cubed')

ax.legend(loc = 0) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

# também dá para colocar loc com coordenadas, ex: (0.1, 0.1)
```

### plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, 
    color = 'orange', 
    lw = 3, 
    alpha = 1, 
    ls = '--', 
    marker = 'o', 
    markersize = 10,
    markerfacecolor = 'yellow',
    markeredgewidth = 2,
    markeredgecolor = 'blue') # linestyles: '--', '-.', ':', 'solid'; markers: 'o', '*', '+', '1'

# limites dos eixos
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
