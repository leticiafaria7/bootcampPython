### distribution plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 40)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg') 
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde') # kde = densidade de probabilidade?
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

### categorical plots

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex') # split = True does not work
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# does not work anymore
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar') 
```

### matrix plots

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
fp
```

```python
sns.heatmap(fp, cmap = 'coolwarm', linecolor = 'white', linewidths = 1)
```

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale = 1)
```

### grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
